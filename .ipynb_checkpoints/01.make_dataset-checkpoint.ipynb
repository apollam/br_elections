# Get RJ

```python
def get_candidates_by(df, cargo, uf):
    df = df[(df['descricao_cargo']==cargo) & (df['sigla_uf']==uf)]
    chosen_cols = ["ano_eleicao", "nome", "nome_urna", "numero_sequencial", "numero_urna", "sigla_uf_nascimento", "sigla_partido", 
                "tipo_agremiacao", "descricao_cargo", "despesa_maxima_campanha", "idade_data_posse", 
                "descricao_cor_raca", "descricao_estado_civil", "descricao_genero", "descricao_grau_instrucao", 
                "descricao_ocupacao", "concorre_reeleicao", "declara_bens",
                "descricao_totalizacao_turno"]
    df = df[chosen_cols]
    return df 
```

```python
cand_rj = get_candidates_by(cand, 'DEPUTADO FEDERAL', 'RJ')
```
